# Data Wrangling

```python
import pandas as pd

df = pd.read_csv('/content/cars sales.csv')
```

```python
print(df.head())
```

```python
print(df.describe(include = 'all'))
```

```python
print(df.dtypes)
```

## Handling Missing Values

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace = True)
```

```python
print(df.isnull().sum())
```

## Handle date column

```python
from dateutil import parser

def parse_date(date_str):
  try:
    return parser.parse(date_str, dayfirst = True)
  except ValueError:
    return parser.parse(date_str)

df['Date'] = df['Date'].apply(parse_date)
```

```python
df
```

## Convert Data Types

```python
df['Annual Income'] = pd.to_numeric(df['Annual Income'], errors = 'coerce')
df['Price ($)'] = pd.to_numeric(df['Price ($)'],errors = 'coerce')
```

```python
df
```

```python
print(df.dtypes)
```

## Convert car id column

```python
df['Car_id'] = df['Car_id'].str.extract('(\d+)$').astype(int)
```

```python
print(df.head())
```

```python
df
```

## Remove Duplicates from data


```python
df.drop_duplicates(inplace = True)
```

## Save cleaned sales data in 'Cleaned Car Sales'

```python
df.to_csv('Cleaned Car Sales.csv', index=False)
```
